### Introducción

En el presente trabajo, se realiza un análisis estadístico detallado de los retornos diarios de las acciones de YPF, con el objetivo de evaluar su rendimiento, volatilidad y riesgos asociados. Este estudio se fundamenta en métricas clave como la media, la desviación estándar, el coeficiente de variación y la distribución de frecuencias de los retornos diarios, complementado con pruebas de bondad de ajuste y probabilidades asociadas a eventos extremos.

El análisis busca ofrecer una visión integral de las características del activo, identificando tanto las oportunidades como las limitaciones para distintos tipos de inversores. En particular, se examina el equilibrio entre el rendimiento promedio y el alto nivel de volatilidad, así como las implicaciones prácticas para la gestión de riesgos y la toma de decisiones de inversión.

Además, se exploran aspectos relacionados con la adecuación de la distribución normal para modelar los retornos, evaluando la necesidad de considerar alternativas que capturen mejor las características observadas en los datos. Finalmente, se discuten las posibles estrategias de inversión en el corto y largo plazo, con base en los hallazgos obtenidos.

```python
# Instalar yfinance si no está instalado
!pip install yfinance --quiet

# Importar bibliotecas necesarias
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Descargar datos históricos de YPF
data = yf.download("YPF", start="2020-01-01", end="2023-12-31")

# Exploración de los datos
data_close = data['Close']  # Usaremos los precios de cierre

# Calcular retornos diarios (logarítmicos para análisis financiero)
returns = data_close.pct_change().dropna()

# Estadísticas descriptivas
mean_return = returns.mean()  # Media
std_dev = returns.std()  # Desviación estándar
coef_variation = std_dev / mean_return  # Coeficiente de variación

# Gráfico: Histograma de retornos
plt.figure(figsize=(10, 6))
plt.hist(returns, bins=50, color='blue', alpha=0.7)
plt.title("Histograma de Retornos Diarios de YPF")
plt.xlabel("Retorno Diario")
plt.ylabel("Frecuencia")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Gráfico: Tendencia de precios de cierre
plt.figure(figsize=(10, 6))
plt.plot(data_close, label="Precio de Cierre", color='orange')
plt.title("Tendencia de Precios de Cierre de YPF (2020-2023)")
plt.xlabel("Fecha")
plt.ylabel("Precio de Cierre")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Mostrar resultados
print("Media de Retornos Diarios:", mean_return)
print("Desviación Estándar de Retornos Diarios:", std_dev)
print("Coeficiente de Variación:", coef_variation)
```

Con los datos proporcionados de la primera parte, podemos sacar las siguientes conclusiones específicas sobre los retornos diarios de YPF:

---

### **1. Media de los Retornos Diarios**  
- **Media**: 0.001294 (0.1294%)  
  Esto indica que, en promedio, el rendimiento diario de las acciones de YPF es positivo, aunque modesto.  
  - Si este promedio se mantiene constante, implica un crecimiento acumulativo en el largo plazo. Por ejemplo, si este rendimiento promedio se anualiza (suponiendo 252 días de mercado al año), resultaría en aproximadamente un **rendimiento anual esperado de ~32.67%** (\( (1 + 0.001294)^{252} - 1 \)).
  - A corto plazo, el bajo rendimiento diario implica que las oportunidades de ganancia provienen más de la variabilidad (volatilidad) que de una tendencia sostenida.

---

### **2. Desviación Estándar de los Retornos Diarios**  
- **Desviación estándar**: 0.041942 (4.1942%)  
  Esto refleja la volatilidad de los rendimientos diarios de YPF.  
  - Un movimiento típico en el precio de las acciones de YPF podría variar alrededor del ±4.19% de su valor actual en un día promedio. Esto sugiere que el activo tiene una **alta volatilidad**, lo cual es común en acciones relacionadas con mercados emergentes o sectores como energía y petróleo.

---

### **3. Coeficiente de Variación (CV)**  
- **CV**: 32.415301 (~32.42)  
  El coeficiente de variación mide la relación entre la desviación estándar y la media. En este caso, el CV es extremadamente alto, indicando que la magnitud de la volatilidad es mucho mayor en comparación con el rendimiento promedio diario.  
  - Un CV alto (generalmente > 1 o 100%) implica que la inversión tiene **un alto nivel de riesgo relativo en comparación con el retorno esperado**.  
  - Es decir, aunque hay un rendimiento promedio positivo, este se acompaña de una gran incertidumbre, lo que podría dificultar la previsión de resultados a corto plazo.

---

### **Conclusión General de la Primera Parte**  
1. **Riesgo vs. Retorno**:  
   - YPF ofrece un rendimiento promedio positivo, pero está acompañado de una alta volatilidad relativa, como lo indica el CV.  
   - Los inversores deben estar preparados para enfrentarse a grandes fluctuaciones diarias, lo que lo hace más adecuado para inversores con alta tolerancia al riesgo o traders que buscan aprovechar movimientos significativos.

2. **Perspectiva a Largo Plazo**:  
   - Si el rendimiento promedio diario se mantiene constante, YPF podría ser atractivo para una estrategia de largo plazo. Sin embargo, el alto CV sugiere que el rendimiento acumulado podría variar considerablemente debido a la volatilidad.

3. **Implicación para la Gestión de Riesgos**:  
   - Los inversores deben considerar estrategias para mitigar riesgos, como diversificación o límites de pérdidas (stop-loss), dado el alto nivel de incertidumbre.


```python

!pip install yfinance

# Importar librerías
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, kstest

# Descargar los datos de YPF
data = yf.download("YPF", start="2020-01-01", end="2023-12-31")
data_close = data['Close']

# Calcular retornos diarios
returns = data_close.pct_change().dropna()

# Calcular frecuencia relativa de días positivos y negativos
positive_days = (returns > 0).sum()
negative_days = (returns < 0).sum()
total_days = len(returns)
freq_positive = positive_days / total_days
freq_negative = negative_days / total_days

# Ajuste a distribución normal
mean, std = norm.fit(returns)  # Ajuste de media y desviación estándar
x = np.linspace(returns.min(), returns.max(), 100)
pdf = norm.pdf(x, loc=mean, scale=std)  # Función de densidad ajustada

# Prueba de bondad de ajuste (Kolmogorov-Smirnov)
ks_stat, ks_pvalue = kstest(returns, 'norm', args=(mean, std))

# Probabilidad de que el retorno diario exceda un cierto valor (e.g., 2%)
threshold = 0.02
prob_exceed = 1 - norm.cdf(threshold, loc=mean, scale=std)

# Gráfico: Histograma de retornos con la distribución ajustada
plt.figure(figsize=(10, 6))
plt.hist(returns, bins=50, density=True, color='blue', alpha=0.7, label="Histograma de Retornos")
plt.plot(x, pdf, color='red', linewidth=2, label="Distribución Normal Ajustada")
plt.title("Ajuste de los Retornos Diarios de YPF a una Distribución Normal")
plt.xlabel("Retorno Diario")
plt.ylabel("Densidad de Frecuencia")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Resultados
print({
    "Frecuencia Días Positivos": freq_positive,
    "Frecuencia Días Negativos": freq_negative,
    "Media (Normal)": mean,
    "Desviación Estándar (Normal)": std,
    "Estadístico KS": ks_stat,
    "P-valor KS": ks_pvalue,
    "Probabilidad de Exceder 2%": prob_exceed
})
```

Con base en los resultados específicos, Podemos concluir lo siguiente ⚛

---

### **1. Frecuencia de días positivos y negativos**  
- **Frecuencia de días positivos**: 47.76%  
- **Frecuencia de días negativos**: 49.35%  
La proporción de días positivos y negativos es bastante equilibrada, lo que indica que el comportamiento de los retornos diarios es aproximadamente simétrico. Sin embargo, hay una ligera predominancia de días negativos. Esto sugiere que, en el corto plazo, las acciones de YPF tienen una tendencia neutra a levemente bajista.

---

### **2. Estadísticas de la distribución normal ajustada**  
- **Media de los retornos diarios**: 0.00129 (0.129% diario promedio)  
  Esto indica que, en promedio, los retornos diarios de YPF son ligeramente positivos, aunque cercanos a cero. Si se mantiene este promedio, podría traducirse en un crecimiento acumulado a largo plazo, pero no significativo en el corto plazo.  
- **Desviación estándar**: 0.04192 (4.19%)  
  Esto refleja una alta volatilidad diaria en los precios de las acciones. Los movimientos diarios pueden variar considerablemente, indicando que el activo tiene un riesgo significativo en el corto plazo.

---

### **3. Prueba de bondad de ajuste (Kolmogorov-Smirnov)**  
- **Estadístico KS**: 0.0651  
- **P-valor KS**: 0.00037  
Dado que el p-valor es muy bajo (< 0.05), se rechaza la hipótesis de que los retornos diarios siguen perfectamente una distribución normal. Aunque visualmente pueden parecer similares a una normal, existen desviaciones significativas, posiblemente debido a colas más gruesas o asimetría.

---

### **4. Probabilidad de exceder un retorno diario del 2%**  
- **Probabilidad**: 32.77%  
Hay un 32.77% de probabilidad de que los retornos diarios de YPF excedan el 2% en un día dado, ya sea positiva o negativamente. Esto refleja la naturaleza altamente volátil del activo, con movimientos significativos no siendo eventos raros.

---

### **Conclusiones prácticas**
1. **Oportunidades de inversión**:  
   - La media positiva sugiere que, en el largo plazo, las acciones de YPF podrían ofrecer rendimientos modestos, aunque la volatilidad podría desincentivar a los inversores más conservadores.
   - La probabilidad alta de exceder ±2% diarios implica oportunidades para traders que busquen capitalizar movimientos grandes en el corto plazo.

2. **Gestión de riesgos**:  
   - La alta desviación estándar y la distribución no perfectamente normal indican que hay un riesgo elevado de movimientos extremos. Los inversores deberían considerar estrategias de cobertura o establecer límites estrictos para gestionar pérdidas.

3. **Ajuste del modelo**:  
   - Aunque los datos se asemejan a una distribución normal, sería útil explorar otras distribuciones (e.g., t-Student o colas gruesas) para modelar mejor los eventos extremos.

4. **Recomendación general**:  
   - Para estrategias a largo plazo, los rendimientos esperados son bajos en comparación con el nivel de volatilidad. Sin embargo, para traders a corto plazo, las oportunidades derivadas de movimientos significativos podrían ser atractivas.
